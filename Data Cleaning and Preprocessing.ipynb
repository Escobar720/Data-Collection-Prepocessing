```python
upper_limit = df['bath'].mean() + 3* df['bath'].std()
lower_limit = df['bath'].mean() - 3* df['bath'].std()
print("upper_limit" , upper_limit)
print("lower_limit" , lower_limit)
```

```python
df.loc[(df['bath'] > upper_limit) | (df['bath'] <lower_limit)]
```

```python
new_df = df.loc[(df['bath'] > upper_limit) | (df['bath'] <lower_limit)]
print("old data" , len(df))
print("new data", len(new_df))
```
